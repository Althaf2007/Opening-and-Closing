Opening and Closing

Name : K.Mohamed Althaf Reg No : 212224240089 Slot : 25EV2130

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = np.zeros((500,1000), dtype=np.uint8)
```

```python
cv2.putText(img, "MOHAMED ALTHAF", (50,250),
            cv2.FONT_HERSHEY_SIMPLEX, 3, 255, 6)
```

```python
kernel = np.ones((3,3), np.uint8)
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(figsize=(18,8))

plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title("Original Text", fontsize=16)
plt.axis('off')

plt.subplot(132)
plt.imshow(opening, cmap='gray')
plt.title("Opening", fontsize=16)
plt.axis('off')

plt.subplot(133)
plt.imshow(closing, cmap='gray')
plt.title("Closing", fontsize=16)
plt.axis('off')

plt.tight_layout()
plt.show()
```
